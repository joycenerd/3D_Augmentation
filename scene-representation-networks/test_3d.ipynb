```python
def pred_2_rgb(img):
    img += 1.
    img /= 2.
    img *= 2**8 - 1
    img = img.round().clip(0, 2**8-1)
    return img
```

```python
def read_pcd(mode, instance_id, pose_id):
#     content = np.load("./3D_points/car_200/cars_" + mode + "_" + str(instance_id) + "_" + str(pose_id).zfill(6)+ "_points.npy")
    content = np.load("./3D_points/shapenet.v2_airplane_car_chair_interpolation/" + str(instance_id) + "_" + str(pose_id).zfill(6)+ "_points.npy")
    content = content[0]
    
#     color_rgb = np.load("./3D_points/car_200/cars_" + mode + "_" + str(instance_id) + "_" + str(pose_id).zfill(6)+ "_rgb.npy")
    color_rgb = np.load("./3D_points/shapenet.v2_airplane_car_chair_interpolation/"+ str(instance_id) + "_" + str(pose_id).zfill(6)+ "_rgb.npy")
    color_rgb = color_rgb[0]
    img = pred_2_rgb(color_rgb)
    
#     filt = np.arange(16384)
    filt = np.where(np.average(img, axis=1) <= 230)
    
    return content[filt], img[filt]
```

```python
from os import listdir
def create_class_list():
    airplane_list = listdir("/eva_data/psa/NVS/datasets/ShapeNet.v2_128/airplane_train")
    car_list = listdir("/eva_data/psa/NVS/datasets/ShapeNet.v2_128/car_train")
    chair_list = listdir("/eva_data/psa/NVS/datasets/ShapeNet.v2_128/chair_train")
    return airplane_list, car_list, chair_list
```

```python
import open3d as o3d
def write_pcd(point, output_path):
    # Convert numpy array to pcd format
    pcd = o3d.geometry.PointCloud()
    pcd.points = o3d.utility.Vector3dVector(point)

    # Output pcd file
    o3d.io.write_point_cloud(output_path, pcd)
```

```python
import random
import numpy as np 
mode = "train_test"
def save_point_cloud():
    air_list, car_list, chair_list = create_class_list()
    air_idx, car_idx, chair_idx = 0, 0, 0
    id_list = sorted(listdir("/eva_data/psa/NVS/datasets/ShapeNet.v2_128/airplane_car_chair_train"))
    for i in range(600):
        print(i)
        random_view = random.sample(range(50), 5)
        points = []
        print(random_view)
        for view in random_view:
            content, _ = read_pcd(mode, str(i), str(view))
            points.extend(content)
        
        output_root = "/eva_data/psa/Augmentation/datasets/airplane_car_chair/"
        if id_list[i] in air_list:
            output_path = output_root + "airplane/test/airplane_" + str(air_idx).zfill(4) + ".pcd"
            air_idx += 1
        elif id_list[i] in car_list:
            output_path = output_root + "car/test/car_" + str(car_idx).zfill(4) + ".pcd"
            car_idx += 1
        elif id_list[i] in chair_list:
            output_path = output_root + "chair/test/chair_" + str(chair_idx).zfill(4) + ".pcd"
            chair_idx += 1
        else:
            print("Something went wrong")
        write_pcd(np.array(points), output_path)
save_point_cloud()
```
